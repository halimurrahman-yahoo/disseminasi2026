```python
%%bash

SRC="/content/drive/MyDrive/MSWEP_V316_test/Past/3hourly"
DST="/content/drive/MyDrive/MSWEP_17_18Juni2016/3hourly"

mkdir -p "$DST"

find "$SRC" -maxdepth 1 \
    \( -name "2016169.*.nc" -o -name "2016170.*.nc" \) \
    -print0 | while IFS= read -r -d '' file
do
    rsync -ah --progress "$file" "$DST/"
done
```
